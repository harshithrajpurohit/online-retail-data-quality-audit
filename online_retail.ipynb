```python
df.groupby('Country')['Revenue'].sum().sort_values(ascending=False)
```

```python
clean_df = df.copy()

clean_df = clean_df[~clean_df['InvoiceNo'].astype(str).str.startswith('C')]
clean_df = clean_df[clean_df['Quantity'] > 0]
clean_df = clean_df[clean_df['UnitPrice'] > 0]
clean_df = clean_df.drop_duplicates()

clean_df.shape
```

```python
clean_df['Revenue'] = clean_df['Quantity'] * clean_df['UnitPrice']
clean_revenue = clean_df['Revenue'].sum()
clean_revenue
```

```python
raw_revenue, clean_revenue, raw_revenue - clean_revenue
```

clean revenue is more because of removal of -ve unit price, -ve quantity(returns) and cancelled orders
